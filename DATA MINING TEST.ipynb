# Section 3- Data Visualization
Analysis and charts on the distributions of Taxi-out time across States,Cities and Airlines as it affect fuel consumption 

```python
# using the data for first half of 2019
half_2019_df.head()
```

```python
taxi_out_time_analysis = half_2019_df.loc[:,['OP_UNIQUE_CARRIER','ORIGIN_CITY_NAME','ORIGIN_STATE_NM','TAXI_OUT']]  
taxi_out_time_analysis.head()
```

### Analysis
Firstly we perform statistical test analysis to check
whether the __average Taxi-out time across Cities and State is similar or not__ using the __ANOVA__ test

__ANOVA test statistics__--> This is analysis of variance test that can be thought of as a generalization of t-tests for more than two groups. Where the independent of t-test is used to compare the means of a condition between more than two groups.Performing ANOVA requires setting a Null and Alternate hypothesis.

- Null hypothesis: There is no significant difference among the groups.
- Alternate Hypothesis: There is a significant difference among the groups.
Therefore, these hypothesis are evaluated based on p-value associated with F-statistics. When p-value is < 0.05, then we reject the Null hypothesis and accept the alternative hypothesis. Also when the p-value is > 0.05 we reject the alternative hypothesis and accept the null hypothesis,thereby going with the conclusion that the means of the groups are equal,hence there is no significant difference

___Test Condition__:

- __H0__: Average Taxi-out time across Cities or States is similar 
- __H1__: Not similar

```python
# impport neccessary libraries for statistical analyses
import scipy.stats as stats
from math import sqrt
```

```python
#For a quick generalization, we extract the first six features for analysis

top_5_city = taxi_out_time_analysis['ORIGIN_CITY_NAME'].value_counts()[:5]

# fetch the names of the top 5 city to be used for analysis
top_5_city_names = top_5_city.index
top_5_city_names
```

```python
#create sample test with the above features
sample_test = taxi_out_time_analysis.loc[taxi_out_time_analysis['ORIGIN_CITY_NAME'].isin(top_5_city_names),['ORIGIN_CITY_NAME','TAXI_OUT']]
sample_test.head()
```

```python
# drop any form of missing values in our sample data
sample_test.dropna(how='any', inplace=True)
```

```python
# creating the individual five sample space from our sample test data
s1 = sample_test[sample_test['ORIGIN_CITY_NAME']== top_5_city_names[0]].TAXI_OUT
s2 = sample_test[sample_test['ORIGIN_CITY_NAME']== top_5_city_names[1]].TAXI_OUT
s3 = sample_test[sample_test['ORIGIN_CITY_NAME']== top_5_city_names[2]].TAXI_OUT
s4 = sample_test[sample_test['ORIGIN_CITY_NAME']== top_5_city_names[3]].TAXI_OUT
s5 = sample_test[sample_test['ORIGIN_CITY_NAME']== top_5_city_names[4]].TAXI_OUT
```

```python
# test statistics
stats.f_oneway(s1,s2,s3,s4,s5)
```

The __ANOVA__ test result shows the __p-value__ is less than __0.05__, so therefore reject null hypothesis and conclude the avaerage Taxi-out time across the different cities and States are not the same
